```python
import cv2 # Open CV - computer vision
import numpy as np
import os # Not only for filepaths but for allowing delays so software can have time to detect different angles
import time
import uuid # For naming image files
from PIL import Image # Pillow library - for imaging & manipulation
from PIL import ImageGrab
from pytesseract import * # Pytesseract - optical character recognition (OCR) tool. It will recognize/“read” image text
```

### Method 1

```python
image_file = "Image_Samples/Cover.png"
im = Image.open(image_file)
text = image_to_string(im)
text = image_to_string(image_file)
#text = image_to_string(image_file, graceful_errors=True)
print("=====output=======\n")
print(text)
```

### Method 2

```python
img = cv2.imread(image_file)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Detecting Characters 
hImg, wImg,_ = img.shape
boxes = pytesseract.image_to_boxes(img)
for b in boxes.splitlines():
    print(b)
    b = b.split(' ')
    print(b)
    x, y, w, h = int(b[1]), int(b[2]), int(b[3]), int(b[4])
    cv2.rectangle(img, (x,hImg- y), (w,hImg- h), (50, 50, 255), 2)
    cv2.putText(img,b[0],(x,hImg- y+25),cv2.FONT_HERSHEY_SIMPLEX,1,(50,50,255),2)
```

```python
cv2.imshow('img', img)
cv2.waitKey(0)
```
